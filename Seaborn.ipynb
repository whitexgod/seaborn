```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
#distribution graph
#.......> 1. distplot
#.......> 2. jointplot
#.......> 3. pairplot
#.......> 4. rugplot

#categorical plot
#.......> 1.barplot
#.......> 2.countplot
#.......> 3.boxplot
#.......> 4.violinplot

#matrix plot
#.......> 1.heatmap
```

# distribution graph

## 1. Distplot

```python
sns.distplot(tips['total_bill'], kde=False)
plt.show()
```

## 2. Jointplot

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', color='orange')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg', color='green')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', color='red')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips )
plt.show()
```

## 3. Pairplot

```python
sns.pairplot(tips) #works on only numerical data
plt.show()
```

```python
sns.pairplot(tips, hue='sex') #works on only numerical data
plt.show()
```

## 4. Rugplot

```python
sns.rugplot(tips['total_bill'] , color="green")
plt.show()
```

# categorical plot

## 1. Barplot

```python
sns.barplot(x='day', y='tip',data=tips, hue='sex')
```

```python
sns.barplot(x='day', y='tip',data=tips, hue='smoker')
```

```python
sns.barplot(x='day', y='tip',data=tips, hue='sex', estimator=np.median)
plt.show()
```

## 2. Countplot

```python
sns.countplot(tips['smoker'],data=tips,hue='sex')
plt.show()
```

## 3. Boxplot

```python
#boxplot works on only numerical data
sns.boxplot('total_bill',data=tips)
plt.show()
```

```python
sns.boxplot(x=tips['sex'],y=tips['tip'])
plt.show()
```

## 4. Violinplot

```python
sns.violinplot(x=tips['sex'],y=tips['tip'])
plt.show()
```

```python
sns.violinplot(x=tips['day'],y=tips['tip'])
plt.show()
```

## Matrixplot.........> heatmap

```python
flight=sns.load_dataset('flights')
flight.head()
```

```python
p=flight.pivot_table(index='year',columns='month',values='passengers')
p
```

```python
sns.heatmap(p)
```

```python
sns.heatmap(p, annot=True, linewidth=2, linecolor='red', cmap='winter')
plt.show()
```
